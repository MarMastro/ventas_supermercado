```python
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/plotly/datasets/master/supermarket_Sales.csv'

df = pd.read_csv(url, parse_dates = ['Date']) # parse_date convierte Date a datetime
print(df.head())
```

```python
df.info()
```

```python
df.describe()
```

```python
# Metrica clave. Ingreso = quantity * unit price

df['Ingreso'] = df['Quantity'] * df['Unit price']
df.head()
```

```python
df = df.drop('ingreso', axis=1)
```

```python
df.head()
```

```python
# Top 5 productos por ingreso total

top5 = (df.groupby('Product line')['Ingreso']
        .sum()
        .sort_values(ascending=False)
        .head(5))
top5
```

```python
import matplotlib.pyplot as plt

top5 = (df.groupby('Product line')['Ingreso']
        .sum()
        .nlargest(5))  # Se queda con los 5 mas grandes

# plot

plt.figure(figsize = (8, 5))
top5.plot(kind = 'bar')
plt.title('Top 5 productos por Ingreso Total')
plt.xlabel('Producto')
plt.ylabel('Ingreso U$S')
plt.xticks(rotation = 45, ha = 'right')
plt.tight_layout()

plt.show()

```
